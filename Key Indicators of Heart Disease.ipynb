```python
#7) Chi-Squared (X2) test for categorical features

#X2 test: if p-value<α=0.05, reject H0, H1 valid; if p-value>α=0.05, accept H0, reject H1.

#H0: categorical features independent (not correlated) - no association between variables
#H1: categorical features dependent (correlated) - there is some association between variables
```

```python
cat_cols
```

```python
#Let’s make an alternative data frame using the categorical columns only. We are dropping our y target variable 'HeartDisease':

target = df_cat['HeartDisease']
df_cat.drop('HeartDisease',axis=1,inplace=True)
df_cat.sample()
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.preprocessing import LabelEncoder
```

```python
for col in df_cat.columns:
    le = LabelEncoder()
    df_cat[col]= le.fit_transform(df_cat[col])
```

```python
#Now we have the categorical dataframe with entries converted into numerical values via LabelEncoder

df_cat.head()
```

```python
from sklearn.feature_selection import chi2

#X: all categorical values (but 'HeartDisease'), with entries converted into numerical values
#Y: target independent value ('HeartDisease')

X=df_cat  
y=target

chi_scores=chi2(X,y)
chi_scores
```

```python
#Upper array: Chi-square values (the higher the value, the more important the y target variable against the dependent variable)
#Lower array: p-values (the higher the p-value of the dependent variable, it means that such variable is independent, and hence less important against the y target variable).

#Let’s now plot the chi square values (higher the Chi square value, higher the importance):
```

```python
chi_values=pd.Series(chi_scores[0], index=X.columns)
chi_values.sort_values(ascending=False, inplace=True)
chi_values.plot.bar()
```

```python
#Higher chi-square value: the higher the importance of such variable against the y (‘HeartDisease’).
```

```python
#Now let’s plot the p-values (lower the p-value, higher the importance):

p_values=pd.Series(chi_scores[1], index=X.columns)
p_values.sort_values(ascending=False, inplace=True)
p_values.plot.bar()
```

```python
#‘GenHealth’ has a p-value much bigger than 0.05, so we can drop it as such dependent variable accepts H0 (variables independent). 

#Summing up: all categorical variables but ‘GenHealth’ have some relation to the target variable ‘HeartDisease’ (H1 accepted).
```

```python
df_cat.drop('GenHealth',axis=1,inplace=True)
df_cat
```
